# INSUFFISANCE CARDIAQUE

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import SGDClassifier,LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn import svm,preprocessing
from sklearn.datasets import make_classification
from sklearn.preprocessing import MinMaxScaler, PolynomialFeatures, StandardScaler, LabelBinarizer,LabelEncoder
from sklearn.model_selection import learning_curve, train_test_split, GridSearchCV
from sklearn.metrics import f1_score, classification_report, confusion_matrix,accuracy_score
from sklearn.feature_selection import SelectKBest, f_classif
```

## Chargement des donnees

```python
df_test0=pd.read_excel('New_insuffisance_cardiaque.xlsx')
df_test0
```

```python
df_test0=df_test0.set_index('trustii_id')
df_test0
```

```python
df_train=pd.read_excel('New_Machine learning et Insuffisance Cardiaque_2.xlsx')
df_train
```

## Normalisation MinMax()

```python
scaler = preprocessing.MinMaxScaler()
df_train=pd.DataFrame(scaler.fit_transform(df_train.values),columns=df_train.columns)
df_train
```

```python
df_test=pd.DataFrame(scaler.fit_transform(df_test0.values),columns=df_test0.columns)
df_test.index=df_test0.index
df_test
```

## Variables dependantes et variables explixactives

```python
y_train=df_train['DEATH_EVENT'].values 
y_train
```

```python
x_train=df_train.drop(columns='DEATH_EVENT').values
x_train
```

## Subdivision des donnees d'entrainement en sous donnees de test et d'entrainement.

```python
x_train_train, x_train_test, y_train_train, y_train_test = train_test_split(x_train, y_train,test_size = 0.2,random_state=42)
```

## Test de modeles afin de choisir le meilleur

```python
def test_function():
    m1=RandomForestClassifier()
    m2=GradientBoostingClassifier()
    m3=KNeighborsClassifier(n_neighbors=10)
    m4=LogisticRegression(penalty='none',solver='newton-cg')
    m5=DecisionTreeClassifier()
    m6= AdaBoostClassifier()
    m1.fit(x_train_train,y_train_train)
    m2.fit(x_train_train,y_train_train)
    m3.fit(x_train_train,y_train_train)
    m4.fit(x_train_train,y_train_train)
    m5.fit(x_train_train,y_train_train)
    m6.fit(x_train_train,y_train_train)
    print(m1.score(x_train_test,y_train_test))
    print(m2.score(x_train_test,y_train_test))
    print(m3.score(x_train_test,y_train_test))
    print(m4.score(x_train_test,y_train_test))
    print(m5.score(x_train_test,y_train_test))
```

```python
test_function()
```

## Apres plusieurs tests, la fonction RandomForestClassifer() est de loin le modele qui donne de meilleurs scores
